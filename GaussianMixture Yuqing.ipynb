```python
import scipy.io
import time
import numpy as np
import pandas as pd
import matplotlib
import csv
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
data = pd.read_csv('diffusion_map_20_eigen.tsv', sep='\t',header=None)
data2 = np.asarray(data)
```

```python
runtime_all = []
for N in [15,20,25]:
    start = time.time()
    pred = GaussianMixture(n_components=N,random_state=0).fit_predict(data2)
    end = time.time()
    runtime = end-start
    plt.figure()
    plt.scatter(data2[:,0],data2[:,1],c=pred)
    plt.show()
    runtime_all.append(runtime)
```

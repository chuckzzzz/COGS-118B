```python
%pylab inline
import scipy.io
from sklearn.manifold import TSNE #https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html
import pandas as pd
```

```python
%%time
f="sample.mmtx"
t=scipy.io.mmread(f)
t # <class 'scipy.sparse.coo.coo_matrix'>
```

```python
%%time
tsne_results = TSNE(n_components = 3, verbose=1).fit_transform(t)
```

```python
df = pd.DataFrame(tsne_results)
df.to_csv('tsne_3.tsv', sep='\t', header=False, index=False)
```

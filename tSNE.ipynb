```python
%pylab inline
import scipy.io
from sklearn.manifold import TSNE #https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html
```

```python
%%time
f="sample.mmtx"
t=scipy.io.mmread(f)
t # <class 'scipy.sparse.coo.coo_matrix'>
```

```python
%%time
tsne_results = TSNE(verbose=1).fit_transform(t)
```

```python
plt.scatter(tsne_results[:,0], tsne_results[:,1],s=1)
plt.xlabel('TSNE-1')
plt.ylabel('TSNE-2')
```
